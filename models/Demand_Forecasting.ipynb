# Data Cleaning and Preprocessing

```python
supplyChain = supplyChain.drop(columns=['Product Description', 'Customer Email', 
                                        'Product Image', 'Customer Password'])

supplyChain['Customer Lname'] = supplyChain['Customer Lname'].fillna('Unknown')
supplyChain['Customer Zipcode'] = supplyChain['Customer Zipcode'].fillna('Unknown')

supplyChain['order date (DateOrders)'] = pd.to_datetime(supplyChain['order date (DateOrders)'], errors='coerce')
supplyChain['shipping date (DateOrders)'] = pd.to_datetime(supplyChain['shipping date (DateOrders)'], errors='coerce')

supplyChain['Year'] = supplyChain['order date (DateOrders)'].dt.year
supplyChain['Month'] = supplyChain['order date (DateOrders)'].dt.month
supplyChain['YEAR_MONTH']=supplyChain['order date (DateOrders)'].dt.to_period('M')
supplyChain['YEAR_WEEK']=supplyChain['order date (DateOrders)'].dt.to_period('W')
```

```python
supplyChain.head()
```

```python
supplyChain.columns
```

```python
# Aggregate order quantities by region
supplyChain_group = supplyChain.groupby(['Year', 'Market', 'Month', 'Order Country',
                                        'Order Region', 'Order State', 'Product Name', 
                                         'Category Name'])
```

```python
supplyChain_group1 = supplyChain_group["Order Item Quantity"].sum().reset_index()
supplyChain_group2 = supplyChain_group["Sales"].sum().reset_index()
supplyChain_group3 = supplyChain_group["Days for shipping (real)"].mean().reset_index()
```

```python
supplyChain_group4 = supplyChain_group["Product Price"].mean().reset_index()
supplyChain_group5 = supplyChain_group["Product Status"].mean().reset_index()
```

```python
supplyChain_df = supplyChain_group1.copy()
supplyChain_df["Sales"] = supplyChain_group2["Sales"]
supplyChain_df["Days for shipping (real)"] = supplyChain_group3["Days for shipping (real)"]
supplyChain_df["Product Price"] = supplyChain_group4["Product Price"]
supplyChain_df["Product Status"] = supplyChain_group5["Product Status"]
```

```python
# Dummy variables
supplyChain_df = pd.get_dummies(supplyChain_df, columns = ['Market', 'Order Country',
                                        'Order Region', 'Order State', 'Product Name', 
                                         'Category Name'], dtype=int)
```

```python
supplyChain_df.head()
```

```python
# get X and y
X = supplyChain_df.loc[:, ~supplyChain_df.columns.isin(["Order Item Quantity"])]
y = supplyChain_df["Order Item Quantity"]
```

```python
# Standardization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X = X.copy()
scaled_X = scaler.fit_transform(scaled_X)
scaled_X = pd.DataFrame(scaled_X, columns = X.columns)
```

```python
# Drop all 0 columns
scaled_X = scaled_X.loc[:, (scaled_X!=0).any(axis=0)]
```

```python
# Feature selection
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=695)
rfr.fit(scaled_X, y)
```

```python
selected_ind = np.argsort(rfr.feature_importances_)[1404:]
selected_features = scaled_X.columns[selected_ind]
selected_features
```

```python
X_selected = scaled_X[selected_features]
X_selected.head()
```

```python
# Get the train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size = 0.20, random_state = 695)
```
